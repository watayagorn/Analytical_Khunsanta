```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file into a pandas DataFrame
file_path = "AI analyze_211123.xlsx"  # Replace with the actual path to your Excel file
df = pd.read_excel(file_path)

# Check if the required columns "ai_score" and "xpert" exist
if "ai_score" not in df.columns or "xpert" not in df.columns:
    print("Error: Columns 'ai_score' and/or 'xpert' not found in the DataFrame.")
else:
    # Filter positive samples based on the "xpert" column
    positive_samples = df[df["xpert"] == 1]["ai_score"]

    # Plot distribution plot of "ai_score"
    plt.figure(figsize=(10, 6))
    sns.histplot(df["ai_score"], kde=True, color='skyblue', label='All Samples')

    # Check if there are positive samples before plotting the mean
    if not positive_samples.empty:

        # Mark positive samples on the histogram
        plt.axvline(x=positive_samples.mean(), color='green', linestyle='--', label='Mean of Positive Samples')
        plt.scatter(positive_samples, [0] * len(positive_samples), color='red', s=130, marker='|', label='Positive Samples')

    # Plot settings
    plt.xlabel('AI Score')
    plt.ylabel('Frequency')
    plt.title('Distribution Plot of AI Score with Positive Samples')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
